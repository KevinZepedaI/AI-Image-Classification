Paso 3:  Preparar nuestras imágenes para que la Inteligencia Artificial lo aprenda


````python
# Directorio raíz donde están almacenados los datos.
data_path = "dataset"

# Configurar un DataBlock para manejar los datos.
db = DataBlock(
    # Define que los datos son imágenes y etiquetas categóricas.
    blocks=(ImageBlock, CategoryBlock),
    # Función para obtener todas las imágenes del directorio.
    get_items=get_image_files,
    # Divide los datos en entrenamiento (80%) y validación (20%).
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    # Obtiene las etiquetas a partir del nombre del directorio padre.
    get_y=parent_label,
    # Redimensiona todas las imágenes a 128x128 píxeles.
    item_tfms=Resize(128)
)
````

````python
# Crea dataloaders (manejo de lotes) a partir de los datos configurados.
dls = db.dataloaders(data_path)

# Muestra un lote de datos de validación para verificar.
dls.valid.show_batch(max_n=4, nrows=1)
````

````python
# Crear nuestros datos y crear más de estos
db = db.new(item_tfms = RandomResizedCrop(224, min_scale = 0.5))

# Recortes aleatorios y escalas dinámicas.
dls = db.dataloaders(data_path)
````

````python
# Muestra un lote de datos de entrenamiento con transformaciones aplicadas.
dls.train.show_batch(max_n=8, nrows=2, unique=True)
````

Paso 4: Entrenar nuestro modelo

````python
# Crear un modelo usando ResNet18 y métrica de error_rate (tasa de error).
learn = vision_learner(dls, resnet18, metrics = error_rate)
# Ajustar el modelo (fine-tune) durante 4 épocas.
learn.fine_tune(4)
````

````python
# Exportar el modelo entrenado a un archivo.
learn.export()

# Verificar los archivos disponibles en el directorio actual.
path = Path()

# Lista archivos con extensiones específicas.
path.ls(file_exts = " .")
````

Paso 5: Poner nuestro modelo en marcha

````python
# Importar Gradio para crear una interfaz de usuario.
import gradio as gr
from fastai.vision.all import *
# (Otras importaciones potencialmente necesarias.)
import skimage
# Módulo para manejar el sistema de archivos.
import os
````

````python
# Cargar el modelo entrenado desde el archivo exportado.
learn = load_learner("export.pkl")
````

````python
# Obtener las etiquetas de las clases aprendidas por el modelo.
labels = learn.dls.vocab

# Función para realizar predicciones con el modelo.
def predict(img):
  # Convierte la imagen de entrada a un formato compatible.
    img = PILImage.create(img)
    # Predicción, índice y probabilidades de cada clase.
    pred, pred_idx, probs = learn.predict(img)
    # Devuelve las probabilidades de cada clase.
    return {labels[i]: float(probs[i]) for i in range(len(labels))}
````

````python
# Configuración de la interfaz de usuario con Gradio.

# Título de la interfaz.
title = "Clasificador de Chihuahua o Panqués"
# Descripción breve.
description = "Un clasificador entrenado con imágenes de Chihuahuas  y Panqués."
# Enlace adicional.
article = "<p style='text-align: center'><a href='https://tmabraham.github.io/blog/gradio_hf_spaces_tutorial' target='_blank'>Blog post</a></p>"
interpretation = "default"
enable_queue = True
````

````python
# Crear y lanzar la interfaz interactiva.
gr.Interface(
             # Función para realizar predicciones.
             fn = predict,
             # Entrada: imagen cargada.
             inputs = gr.Image(type = "pil"),
              # Salida: etiquetas con sus probabilidades.
             outputs = gr.Label(num_top_classes = 3),
             title = title,
             description = description,
             article = article,
             ).queue().launch()
````
